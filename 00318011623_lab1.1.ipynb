```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("homeprices - homeprices.csv")
print(df)
```

```python
def simple_linear_regression(x, y):
    n = len(x)

    sum_x = np.sum(x)
    sum_y = np.sum(y)
    sum_xy = np.sum(x * y)
    sum_x_squared = np.sum(x ** 2)

    m = (n * sum_xy - sum_x * sum_y) / (n * sum_x_squared - sum_x ** 2)
    c = (sum_y - m * sum_x) / n

    return m, c
```

```python
def predict(x, m, c):
    return m * x + c
```

```python
X = df['area'].values
Y = df['price'].values

slope, intercept = simple_linear_regression(X, Y)
predicted = predict(X, slope, intercept)

print(f"Slope : ", {slope})
print(f"Intercept : ", {intercept})
print(predicted)
```

```python
plt.plot(X, predicted, color = 'blue', label = 'Regression Line')
plt.scatter(X, Y, color = 'red', label = 'Scatter Plots')
plt.xlabel('area')
plt.ylabel('price')
plt.legend()
```
